```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data ={
    "Pilot ID":["P128","P145","P104","P102","P100","P122"],
    "Name":["Jane Doe","Ryan Mak","Lionel Larry","Amy Annie","Tina Travis","Laura Mak"],
    "Hours Flown":[200,3500,280,300,267,280],
    "Type of flight":["Domestic","International","Domestic","International","Domestic","International"],
    "Flights Completed":[20,125,50,29,35,76]
}
```

```python
df= pd.DataFrame(Data)
```

```python
# Set plot style
sns.set(style="whitegrid")
```

```python
# Create a figure for plotting
plt.figure(figsize=(10, 6))
```

```python
# Plot Hours Flown by Flight Type
sns.barplot(x="Name", y="Hours Flown", data=df, hue="Type of flight", palette="Set2")
# Rotate the x-axis labels to prevent overlap
plt.xticks(rotation=45, ha="right")
plt.title('Pilot Hours Flown by Flight Type', fontsize=14)
plt.xlabel('Pilot', fontsize=12)
plt.ylabel('Hours Flown', fontsize=12)
```

```python

# Display the plot
plt.tight_layout()
plt.show()
```
