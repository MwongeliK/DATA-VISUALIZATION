```python
import pandas as pd
import numpy as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
height = [150,100,165,185]
weight = [70,80,90,100]
```

```python
plt.plot(height,weight)
```

```python
#Giving the above graph a title
plt.title("Relationship between height and Weight")

#labelling the axes
plt.xlabel("Height")
plt.ylabel("Weight")
plt.plot(height,weight)
```

```python
calories_burnt=[65,75,95,99]
plt.plot(calories_burnt)
plt.plot(weight)


#using legend and loc
plt.legend(labels=['calories burnt','weight'],loc='lower right')

plt.xticks(ticks = [0,1,2,3], labels=['p1','p2','p3','p4']);

```

```python
#using different dimensions on the graph
plt.figure(figsize=(15,5))
plt.plot(calories_burnt)
plt.plot(weight)


#using legend and loc
plt.legend(labels=['calories burnt','weight'],loc='lower right')

plt.xticks(ticks = [0,1,2,3], labels=['p1','p2','p3','p4']);


```

```python
plt.plot(calories_burnt)
#changing colors of the plot
plt.plot(weight,'bo')



#using legend and loc
plt.legend(labels=['calories burnt','weight'],loc='lower right')

plt.xticks(ticks = [0,1,2,3], labels=['p1','p2','p3','p4']);

```

```python
#creating subplots
fig,ax=plt.subplots(nrows = 2,ncols=2, figsize=(5,5))
ax[0,0].plot(calories_burnt,'g--')
ax[0,1].plot(weight,'r-')
ax[1,0].set_xticklabels(labels=['p1','p2','p3','p4'])
```
