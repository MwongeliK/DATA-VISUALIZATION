```python
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
#loading the dataset
task=["Planning","Defining requirements","Wireframe Design","Frontend and backend development","UI/UX design", "Content creation",
      "Testing and Debugging","Deployment"]
Startdate= [ "2024-3-29", "2024-3-30", "2024-3-30","2024-4-2","2024-4-3","2024-4-5","2024-4-7","2024-4-9" ]
EndDate= ["2024-3-30","2024-3-31","2024-4-2","2024-4-4","2024-4-7","2024-4-6","2024-4-9","2024-4-10"]
df=pd.DataFrame(data={"Task":task, "Start":Startdate, "End":EndDate})

df["Start"]=pd.to_datetime(df.Start)
df["End"]=pd.to_datetime(df.End)

df["Days"]=df["End"] - df["Start"]
df["color"]=plt.cm.Set1.colors[:len(df)]
df
```

```python
from datetime import date
plt.style.use("ggplot");

fig = plt.figure(figsize=(10,7))
plt.barh(y=df["Task"], left = df["Start"], width=df["Days"] ,color=df['color'] );
plt.xlim(date(2024,3,28), date(2024,4,10));
plt.ylim(-0.5,7.5);

dt_rng=pd.date_range(start="2024-3-28",end="2024-4-10", freq='D')
plt.xticks(dt_rng, [dt.day_name() for dt in  dt_rng] ,fontsize=16, rotation=30, horizontalalignment= 'right');
plt.yticks(fontsize=16,rotation=30, horizontalalignment= 'right');


plt.xlabel("Days" ,fontsize=20, fontweight="bold")
plt.ylabel("Task" ,fontsize=20 , fontweight="bold")
plt.title("PROJECT ROADMAP" , fontsize=20,fontweight="bold")
```
